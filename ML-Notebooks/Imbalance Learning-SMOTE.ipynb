```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/car_evaluation.csv')
df.head()
```

```python
df.shape
```

```python
df.outcome.value_counts()
```

```python
for col in df.columns:
  print(df[col].value_counts())
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['buying'] = le.fit_transform(df.buying)
df['maint'] = le.fit_transform(df.maint)
df['doors'] = le.fit_transform(df.doors)
df['persons'] = le.fit_transform(df.persons)
df['lug_boot'] = le.fit_transform(df.lug_boot)
df['safety'] = le.fit_transform(df.safety)
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
X.head()
```

```python
y= df.outcome
y.head()
```

```python
# Train test split

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=47)
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
y_train.value_counts()
```

```python
# model building

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred_train = model.predict(x_train)
```

```python
# model evaluation

from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
accuracy_score(y_train,y_pred_train)
```

### BALANCING TECHNIQUE

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
```

```python
x_train_smote, y_train_smote = smote.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
y_train_smote.value_counts()
```

```python
# model building after balancing

model1 = KNeighborsClassifier() # for balanced data
```

```python
model1.fit(x_train_smote,y_train_smote)
```

```python
y_pred1 = model1.predict(x_test)
```

```python
y_pred1_train = model1.predict(x_train_smote)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
accuracy_score(y_train_smote,y_pred1_train)
```

```python
pd.crosstab(y_test,y_pred1)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
# pie chart

data = [2,6,5,10]
plt.pie(data)
plt.show()
```

```python
plt.pie(data,[0,0.1,0.1,0.1])
plt.show()
```

```python
plt.pie(data,[0,0.1,0,0] ,labels = ['t1','t2','t3','t4'])
```

```python
plt.pie(data,[0,0.1,0,0] ,labels = ['t1','t2','t3','t4'],autopct= "%1.1f%%")
```

```python
plt.pie(data,[0,0.1,0,0] ,labels = ['t1','t2','t3','t4'],autopct= "%1.1f%%",shadow=True,colors=['yellow','black','red','orange'])
plt.title('Pie Diagram')
plt.show()
```

## DRAW MULTIPLE GRAPHS IN ONE SINGLE FIGURE (SUBPLOTS)

```python
fig,ax = plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].pie(data)
ax[1,0].bar(x,y)
ax[1,1].scatter(x,y)
```

```python
rcParams['figure.figsize'] = 8,8
fig,ax = plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].pie(data)
ax[1,0].bar(x,y)
ax[1,1].scatter(x,y)
```

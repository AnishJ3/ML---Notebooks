### Face Detection

```python
img = cv2.imread('/content/phant.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
#haar_cas = '/content/'
haar_cas = '/content/phant.jpeg'
```

```python
# reading the XML file

face_cas = cv2.CascadeClassifier(haar_cas)
face_cas
```

```python
# detect the face

faces = face_cas.detectMultiScale(img , scaleFactor = 1.1, minNeighbors=5)
faces
```

```python
# printing the result

print('{} faces detected!!'.format(len(faces)))

for( x, y , w, h) in faces:
  cv2.rectangle(img , (x,y) , (x+w , y + h), (0,200,150) , 5)
```

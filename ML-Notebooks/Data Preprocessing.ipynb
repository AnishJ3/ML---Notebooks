## Outlier removal by replacement with median

```python
q1 = df.bmi.quantile(0.25) # first quartile
q3 = df.bmi.quantile(0.75) # second quartile
```

```python
print(q1)
print(q3)
```

```python
IQR = q3- q1
IQR
```

```python
upper_limit = q3+ 1.5*IQR
upper_limit
```

```python
lower_limit = q1- 1.5* IQR
lower_limit
```

```python
df.median()
```

```python
# Now we are going to rewrite bmi column wherever the values are above upper limit we replace them by median
```

```python
df['bmi'] = np.where(df['bmi']>upper_limit,30,df['bmi'])
```

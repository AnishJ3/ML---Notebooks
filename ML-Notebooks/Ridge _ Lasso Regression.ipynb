```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/50_Startups.csv')
```

```python
df
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.State = le.fit_transform(df.State)
```

```python
df.head()
```

```python
# X and Y drop

X = df.drop(columns=['Profit'],axis=1)
X
```

```python
Y = df.Profit
```

```python
Y
```

```python
from sklearn.preprocessing import MinMaxScaler
scale= MinMaxScaler()
```

```python
X_Scaled = pd.DataFrame(scale.fit_transform(X),columns = X.columns)
X_Scaled.head()
```

```python
# Train test split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_Scaled,Y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

### Model Building

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
r = Ridge()
l = Lasso()
```

```python
r.fit(x_train,y_train)
```

```python
l.fit(x_train,y_train)
```

```python
pred1= r.predict(x_test)
```

```python
pred1
```

```python
pred1_train = r.predict(x_train)
```

```python
pred1_train
```

```python
pred2 = l.predict(x_test)
```

```python
pred2_train =  l.predict(x_train)
pred2_train
```

```python
profit = pd.DataFrame({'Actual Profit' : y_test, 'Ridge Prediction':pred1,'Lasso Prediction':pred2})
profit.head(10)
```

```python
# Checking the accuracy of the Lasso and Ridge Models

from sklearn import metrics
```

```python
# R2 Score

print(metrics.r2_score(y_test,pred1))
print(metrics.r2_score(y_test,pred2))
```

```python
# MSE Score

print(metrics.mean_squared_error(y_test,pred1))
print(metrics.mean_squared_error(y_test,pred2))
```

```python
# RMSE Score

print(np.sqrt(metrics.mean_squared_error(y_test,pred1)))
print(np.sqrt(metrics.mean_squared_error(y_test,pred2)))
```

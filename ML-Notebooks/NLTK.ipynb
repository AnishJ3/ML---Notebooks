```python
# Import required libraries

import pandas as pd
import numpy as np
import nltk

# nltk - natural language toolkit

import regex as re

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

# Text processing

```python
 a= "Hii all!!!! , I am a witcher and nice to meet you all"
```

```python
# Remove all the special characters using Regex
```

```python
reg = re.sub('[^a-zA-Z]',' ',a)
reg
```

```python
# step 2 - make all characters either in lowercase or uppercase
```

```python
lw = reg.lower()
```

```python
lw
```

```python
# step 3: split the input

wd = lw.split()
wd
```

```python
print(stopwords.words('english'))
```

```python
# step 4 : removing stopwords

new = [ i for i in wd if i not in set(stopwords.words('english')) ]
new
```

```python
# list comprehension
# syntax: [Expression/ output for interation in variables]
```

```python
# step 5: stemming - short form of original words

ps = PorterStemmer()
```

```python
p = [ps.stem(w) for w in new]
p
```

```python
# step 6: joining the words

t = ' '.join(p)
t
```

```python
# step 7: convert text to numerical format

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

cv.fit_transform([t]).toarray()
```

```python
df = pd.read_csv('/content/Restaurant_Reviews.tsv',delimiter='\t')
```

```python
df.head()
```

```python
df.shape
```

```python
data = [] # list

for i in range(0,1000):
  review = df['Review'][i] # reading the data
  review = re.sub('^[a-zA-Z]',' ',review) # removing special characters
  review = review.lower()
  review = review.split()
  review = [ ps.stem(w) for w in review if w not in set(stopwords.words('english'))] # stemming and removing stopwords
  review = ''.join(review) # joining all the words
  data.append(review) # for saving the result of the iteration
```

```python
data
```

```python
x = cv.fit_transform(data).toarray()
x
```

```python
y = df['Liked'].values
```

```python
# splitting data

from sklearn.model_selection import train_test_split

X_train, X_test , y_train,y_test = train_test_split(x,y,test_size= 0.3,random_state = 42)
```

### Model Building

```python
from sklearn.naive_bayes import BernoulliNB

model = BernoulliNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Evaluting the model

from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
pd.crosstab(y_test,y_pred)
```

```python
import pickle

pickle.dump(model , open("review.pkl",'wb'))
```

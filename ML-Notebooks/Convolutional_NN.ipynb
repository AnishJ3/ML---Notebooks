### CNN Model Building

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Convolutional2D,MaxPooling2D,Flatten,Dense
```

```python
# adding layers

model = Sequential()

model.add(Convolution2D(32,(3,3) , activation= 'relu' , input_shape = (64,64,3)))  # convolution layer
model.add(MaxPooling2D(pool_size = (2,2)))  # maxpooling layer
model.add(Flatten()) # flatten layer

# we can some hidden layers

model.add(Dense(300,activation = 'relu')) # hidden layer 1
model.add(Dense(150,activation = 'relu')) # hidden layer 2
model.add(Dense(4,activation = 'softmax')) # output layer ( only 4 output layers )
```

```python
# compile the model

model.compile(optimizer = 'adam' ,loss='categorical_crossentropy',metrics = ['accuracy'])
```

```python
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps = [6745,8765,11456,4567,8799,3456,11567]
lwlsteps = [8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
```

```python
plt.plot(days,lwlsteps,'o-b')
```

```python
plt.title("Steps Count - Last Week & This Week")
```

```python
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps = [6745,8765,11456,4567,8799,3456,11567]
lwlsteps = [8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
plt.plot(days,lwlsteps,'o-b')
plt.title("Steps Count - Last Week & This Week")
plt.xlabel("Days of the week")
plt.ylabel("Steps walked this week")
```

```python
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps = [6745,8765,11456,4567,8799,3456,11567]
lwlsteps = [8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
plt.plot(days,lwlsteps,'o-b')
plt.title("Steps Count - Last Week & This Week")
plt.xlabel("Days of the week")
plt.ylabel("Steps walked this week")
plt.grid(True)
```

```python
days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
steps = [6745,8765,11456,4567,8799,3456,11567]
lwlsteps = [8745,6784,11457,5964,7845,7945,13457]
plt.plot(days,steps,'o-g')
plt.plot(days,lwlsteps,'o-b')
plt.title("Steps Count - Last Week & This Week")
plt.xlabel("Days of the week")
plt.ylabel("Steps walked this week")
plt.grid(True)
plt.legend(['Current week','Last week'])
```

```python
# Bar graph

plt.figure(figsize= (10,10)) # height and width of the figure

age = [34,36,41,43,28]
name = ["VK","SKY","MSD","ST","HP"]
plt.bar(name,age)
```

```python
plt.figure(figsize= (10,10))
name = ['Shivam','Suryansh','Shrikant','Vivek']
ot = [29,48,32,46]
ct= [22,34,15,28]
plt.bar(name,ot,label="Open-Tickets",width = 0.5,color='k')
plt.bar(name,ct,label="Closed-Tickets",width = 0.5,color='g')
plt.xlabel("Names")
plt.ylabel('Tickets')
plt.legend()
```

```python
age = [34,36,41,43,28]
name = ["VK","SKY","MSD","ST","HP"]
colors=['g','r','b','k','m']
plt.pie(age,labels=name,colors=colors,startangle = 45)
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/breast_cancer.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.outcome.value_counts()
```

```python
# X and Y split

X = df.iloc[:,:-1]
X.head()
```

```python
y= df.outcome
```

```python
y.head()
```

```python
# train test split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state = 47)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Model building

from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_pred1= model.predict(X_train)
```

```python
y_pred1
```

```python
# Evalutate the model

from sklearn.metrics import accuracy_score,classification_report
```

```python
print("Test accuracy is: ",accuracy_score(y_test,y_pred))
print("Training accuracy is: ",accuracy_score(y_train,y_pred1))
```

```python
pd.crosstab(y_test,y_pred)
```

```python
# Recall

37/(7+37)
```

```python
print(classification_report(y_test,y_pred))
```

### Hyper Parameter Tuning

## GridSearch CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'kernel':['linear','rbf'],
              'C':[0.1,0.2,0.5,1.0],
              'gamma':['scale','auto']}
```

```python
clf = GridSearchCV(model,param_grid = parameters,verbose=2)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
model1 = SVC(C=1,gamma='scale',kernel= 'linear')
```

```python
model1.fit(X_train,y_train)
```

```python
y_=  model1.predict(X_test)
```

```python
print(accuracy_score(y_test,y_))
```

```python
pd.crosstab(y_test,y_)
```
